```python
import pandas as pd
import requests
import time

# Carrega o arquivo Excel
lista_cnpj = pd.read_excel('PET.xlsx')

# Limpa os nomes das colunas (remove espaços extras)
lista_cnpj.columns = lista_cnpj.columns.str.strip()

# Exibe os nomes das colunas para verificar qual contém os CNPJs
print("Colunas disponíveis:", lista_cnpj.columns)

# Substitua 'CNPJ' pelo nome exato da coluna que contém os números
coluna_cnpj = 'CNPJ'  # ajuste conforme necessário

# Lista para armazenar os resultados
df_cnpjs = []

# URL base da API
url_base = 'https://minhareceita.org/'

# Loop para consultar cada CNPJ
for cnpj in lista_cnpj[coluna_cnpj]:
    cnpj = str(cnpj).zfill(14)  # garante que o CNPJ tenha 14 dígitos
    url = url_base + cnpj
    try:
        resposta = requests.get(url)
        if resposta.status_code == 200:
            print(f"CNPJ {cnpj} encontrado com sucesso.")
            df_cnpjs.append(resposta.json())
        else:
            print(f"Erro ao acessar CNPJ {cnpj}: {resposta.status_code}")
    except Exception as e:
        print(f"Erro inesperado com CNPJ {cnpj}: {e}")
    time.sleep(1)  # respeita o tempo entre requisições
```

```python
# Converte os dados coletados em DataFrame
df_resultado = pd.DataFrame(df_cnpjs)

# Salva o arquivo CSV na pasta desejada
df_resultado.to_csv(
    r'C:\Users\igorc\OneDrive\Área de Trabalho\codigok\raspagem_de_dados\resultado_cnpjs.csv',
    index=False,
    encoding='utf-8-sig'
)

print("Arquivo salvo com sucesso em: C:\\Users\\igorc\\OneDrive\\Área de Trabalho\\codigok\\raspagem_de_dados")
```
